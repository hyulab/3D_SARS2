```python
annot_pdb = []
counter = 0

# Set up Original Axes Grid
print "Axes Grid"
axes_max = 75
for i in range(-axes_max, axes_max):
    annot_pdb.append(["ATOM", counter, "C", "", "LIG", "A", 1, "", i, 0, 0, 1, 30, "", "C", "", "*"])
    counter += 1
for i in range(-axes_max, axes_max):
    annot_pdb.append(["ATOM", counter, "C", "", "LIG", "B", 1, "", 0, i, 0, 1, 30, "", "C", "", "*"])
    counter += 1
for i in range(-axes_max, axes_max):
    annot_pdb.append(["ATOM", counter, "C", "", "LIG", "C", 1, "", 0, 0, i, 1, 30, "", "C", "", "*"])
    counter += 1

# Add Plane of separation per Protein
print "Plane 1"
center_A = pdb_df[pdb_df["Chain"] == "A"]["Y"].mean()
for i in tqdm_notebook(range(-axes_max, axes_max, 3)):
    for j in range(-axes_max, axes_max, 3):
        annot_pdb.append(["ATOM", counter, "C", "", "LIG", "D", 1, "", i, center_A, j, 1, 30, "", "C", "", "*"])
        counter += 1
print "Plane 2"
center_B = pdb_df[pdb_df["Chain"] == "B"]["Y"].mean()
for i in range(-axes_max, axes_max, 3):
    for j in range(-axes_max, axes_max, 3):
        annot_pdb.append(["ATOM", counter, "C", "", "LIG", "E", 1, "", i, center_B, j, 1, 30, "", "C", "", "*"])
        counter += 1

# Add rotation circle 1
print "Rot Circle"
circle_height = int(axes_max*0.5)
circle_radius = int(axes_max*0.75)
for i in range(0, 331):
    x = np.cos(np.radians(i))*circle_radius
    z = np.sin(np.radians(i))*circle_radius
    annot_pdb.append(["ATOM", counter, "C", "", "LIG", "F", 1, "", x, circle_height, z, 1, 30, "", "C", "", "*"])
    counter += 1

# Add rotation cone
print "Rot Cone"
for j in range(1, 11):
    circle_height = int(axes_max*0.05*j)
    circle_radius = int(axes_max*0.075*j)
    for i in range(0, 331):
        x = np.cos(np.radians(i))*circle_radius
        z = np.sin(np.radians(i))*circle_radius
        annot_pdb.append(["ATOM", counter, "C", "", "LIG", "G", j, "", x, circle_height, z, 1, 30, "", "C", "", "*"])
        counter += 1

# Add rotation arc 1
print "Rot Arc 1"
arc_height = int(axes_max*0.75)
arc_r1 = axes_max*0.25
arc_r2 = axes_max*0.75
for i in range(0, 181):
    z = 0
    x = np.cos(np.radians(i))*arc_r2
    y = np.sin(np.radians(i))*arc_r1
    annot_pdb.append(["ATOM", counter, "C", "", "LIG", "H", 1, "", x, y + arc_height, z, 1, 30, "", "C", "", "*"])
    counter += 1


# Add rotation arc 2
print "Rot Arc 2"
arc_height = int(axes_max*0.75)
arc_r1 = axes_max*0.25
arc_r2 = axes_max*0.75
for i in range(0, 181):
    x = 0
    z = np.cos(np.radians(i))*arc_r2
    y = np.sin(np.radians(i))*arc_r1
    annot_pdb.append(["ATOM", counter, "C", "", "LIG", "I", 1, "", x, y + arc_height, z, 1, 30, "", "C", "", "*"])
    counter += 1

annot_pdb = pd.DataFrame(annot_pdb, columns = list(pdb_df))

my.df2pdb("Annotations.pdb", annot_pdb)
```

```python
# Reinitialize
pymolCmd.reinitialize()

# Load Structure
pymolCmd.load(dock_oriented, "Viral")
pymolCmd.extract("Human", "Viral and chain B")

# Read Predictions
eclair_preds = pd.read_csv("../../Data/Eclair_Predictions/{0}.txt".format(inter), sep="\t")

# Color By Prediction
cmapA = cm.get_cmap('Blues')   
cmapB = cm.get_cmap('Greens')

for resi, pred in eclair_preds[eclair_preds["Prot"] == 0][["Pos", "Pred"]].values:
    if(np.isnan(pred)):
        pred = 0
    pymolCmd.color(matplotlib.colors.rgb2hex(cmapA(float(pred))[:3]).replace("#", "0x"), "Viral and resi {0}".format(int(resi)))

for resi, pred in eclair_preds[eclair_preds["Prot"] == 1][["Pos", "Pred"]].values:
    if(np.isnan(pred)):
        pred = 0
    pymolCmd.color(matplotlib.colors.rgb2hex(cmapB(float(pred))[:3]).replace("#", "0x"), "Human and resi {0}".format(int(resi)))    

# Set Representation
pymolCmd.hide("everything")
pymolCmd.show("surface")

pymolCmd.load("Annotations.pdb", "X_Grid")
pymolCmd.show("sticks", "X_Grid")
pymolCmd.extract("Y_Grid", "X_Grid and chain B")
pymolCmd.extract("Z_Grid", "X_Grid and chain C")
pymolCmd.extract("Plane_1", "X_Grid and chain D")
pymolCmd.extract("Plane_2", "X_Grid and chain E")
pymolCmd.extract("Rot_1", "X_Grid and chain F")
pymolCmd.extract("Cone", "X_Grid and chain G")
pymolCmd.extract("Arc_1", "X_Grid and chain H")
pymolCmd.extract("Arc_2", "X_Grid and chain I")


# Save
pymolCmd.save("Docking_Steps.pse")
```
